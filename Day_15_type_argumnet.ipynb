# Required Arguments

```python
def add(a,b):
    return a+b

print(add(2,6))
```

# 2 Default arguments

default arguments are those that take default vlaues if no values   is provieded eduring the function cll . if an arguments is not passed  python uses the default value 

```python
def greet(name ,msg="Hello"):
    return f"{msg},{name}!"

print(greet("deepak"))
print(greet("mohit" ,"Goodbye"))
```

## 

```python
def student_info(name ,age,grade):
    return  f"Name : {name}, Age : {age}, Grade : {grade}"

print(student_info(name="Deepak" ,age=20 ,grade="A"))
print(student_info(grade = "B" ,name="Ankit" ,age= 20))
```

## 

```python
def multiply(*args):
    product=  1
    for num in args:
        product = product * num
    return product

print(multiply(2,3))
print(multiply(2,3,4,5))
        
```

## 

```python
def print_info(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Deepak" , age= 20 , city="jaipur" ,course = "Bca")
```
