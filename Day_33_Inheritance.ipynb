# Multiple Inheritance

```python
class Engine:
    def start_engine(self):
        print("Engine started.")

class wheels:
    def roated_wheels(self):
        print("Wheels are rotating")

class car(Engine,wheels):
    def drive(self):
            print("Car is driving")

my_car = car()

my_car.start_engine()

my_car.roated_wheels()

my_car.drive()


              
```

# Multilevel Inheritance

```python
class Animal:

    def eat(self):
        print("animal is eating")

class Dog(Animal):

    def bark(self):
        print("Dog is barking")

class puppy(Dog):

    def weep(self):
        print("Puppy is weeping")

obj1 = puppy()

obj1.eat()
obj1.bark()

obj1.weep()
```

# Case study 1: E-commerce order management


scenario : you are designing an e-commerce order system . the system Includes:

Customer class( with name and contact details )

product class (with name and price)

order class (that inherits customer and product tom mnanage customer orders)

Requiremnents:

customer should store customer details.

product should store product details

order should generate an order summary for a customer ,includeing product details

```python
class Customer:

    def customer_detail(self,name,number):
        self.name = name
        self.number = number


class Product:

    def product_detail(self,pname,price,quan):
        self.pname = pname
        self.price = price
        self.quan = quan


class Order(Customer,Product):
    
    def show(self):
        print(f"customer name is {self.name} and mobile number is {self.number}")
        print(f"product name is {self.pname} and quantity is {self.quan} price is {self.price * self.quan}")


obj = Order()

obj.customer_detail("deepak" ,887569)

obj.product_detail("Tv",4000,2)

obj.show()
    

```
